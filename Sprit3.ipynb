```python
class ScratchLinearRegression_a():    
    def __init__(self, num_iter, lr, bias, verbose):
        # ハイパーパラメータを属性として記録
        self.iter = num_iter
        self.lr = lr
        self.bias = bias
        self.verbose = verbose
        # 損失を記録する配列を用意
        self.loss = np.zeros(self.iter)
        self.val_loss = np.zeros(self.iter)
        
        #シータ
        self.ceet = None

    def _bias(self,X):
        n0=X.shape[0]
        
        if self.bias == True:
            #print("bias True")
            x_one = np.ones(n0)
            x_one_x = np.vstack((x_one,X.T))
            X2 = x_one_x.T
            #print("X2", X2)
        else:
            x_zeros = np.zeros(n0)
            x_zeros_x = np.vstack((x_zeros,X.T))
            X2 = x_zeros_x.T        
        return X2    

                
    def _linear_hypothesis(self,X):
        y1 = self.ceet@X.T
        y_pre = np.reshape(y1, (y1.shape[0],1))
        #print(self.y_pre)
        return y_pre
    
    def _gradient_descent(self, X, y, y_pred):
        # y_pred : (n_samples, 1)
        delta = np.dot((y_pred - y).T, X)
        # delta : (1, n_features)
        delta = np.squeeze(delta.T, axis=1)
        delta /= X.shape[0]
        self.ceet -= self.lr*delta
        
        
        #print(y.shape)
        #print(y1.shape)
        #print(p.shape)
#         print("y[0]", y[0])
#         print("y_pred[0]", y_pred[0])
#         p=(y_pred-y) # y1 : (n_samples, 1), y : (n_samples, 1) -> p : (n_samples, 1)
#         print("p[0]", p[0])
#         tmp = (p.T@X2).T
#         print("tmp[0]", tmp[0])
#         hyp = (p.T@X2)/p.shape[0]
#         hyp = hyp.T[:, 0]
#         print("hyp", hyp)
#         self.ceet -= hyp
# 学習率忘れてた
        
#         yyy = np.multiply(p, X2) # p: (n_samples, 1), X2: (n_samples, 3)
#             # -> yyy : (n_samples, 3)
#         print("yyy.shape", yyy.shape)
#         mean_yyy = yyy.mean(0) # mean_yyy : (3,)
#         print("mean_yyy.shape", mean_yyy.shape)
#         hyp = self.lr*mean_yyy # hyp : (3,)
        
#         print("hyp", hyp)
    
#         self.ceet -= hyp # self.ceet : (3,) -> (3,)
#なぜか極度に大きくなったり小さくなったりして動かなくなる
    
        return self.ceet        

    
    def fit(self, X, y, X_val=None, y_val=None):
        
        X2 = self._bias(X)        
        n1 = X2.shape[1]
        # self.ceet=np.full(n1,random.random())
        self.ceet = np.random.randn(n1)
        
        if self.verbose == 'Ture':
            X2_val = self._bias(X_val)
            
            for i in range(self.iter):
                #print("ceet", self.ceet)
                y_pre=self._linear_hypothesis(X2)
                # print("ハイポのY{}",format(y_pre))
                self.ceet=self._gradient_descent(X2, y, y_pre)
                # print("シータ{}",format(self.ceet))
                # print("繰り返しの回数{}".format(i))
                
                #Trainデータの目的関数(損失関数)
                pp=(y_pre-y)**2
                sum_yyy = pp.sum(0)
                sum_ave_yyy = sum_yyy/(2*y.shape[0])
                np.put(self.loss,[i],sum_ave_yyy)
                
                #訓練データの目的関数(損失関数)
                y_pre_val=self._linear_hypothesis(X2_val)
                
                pp_val=(y_pre_val-y_val)**2
                sum_yyy_val = pp_val.sum(0)
                sum_ave_yyy_val = sum_yyy_val/(2*y_val.shape[0])
                np.put(self.val_loss,[i],sum_ave_yyy_val)
        else:
            for i in range(self.iter):
                y_pre=self._linear_hypothesis(X2)
                # print("ハイポのY{}",format(y_pre))
                self.ceet=self._gradient_descent(X2, y, y_pre)
                # print("シータ{}",format(self.ceet))
                # print("繰り返しの回数{}".format(i))
        
        return(self.ceet,self.loss,self.val_loss)
    
    def predict(self, X):
        
        n0=X.shape[0]
        x_one = np.ones(n0)
        x_one_x = np.vstack((x_one,X.T))
        X2 = x_one_x.T

        y = self.ceet@X2.T
        
        """
        線形回帰を使い推定する。

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            サンプル

        Returns
        -------
            次の形のndarray, shape (n_samples, 1)
            線形回帰による推定結果
        """

        return y
```
