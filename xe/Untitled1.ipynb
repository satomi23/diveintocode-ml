## 栗饅頭が太陽系を覆う時間

```python
import math
taiyou_volume = 4/3*math.pi*(2.50709e+20**3) #太陽系の体積　
kuri_volume = 100 #栗饅頭の体積　
num = 0　#倍になっていく経過時間を格納

while kuri_volume <= taiyou_volume:
   # print("経過： {}分　栗饅頭： {}".format(num,kuri_volume))
    kuri_volume *= 2
    num += 5
else:
    print("栗饅頭が太陽系を覆う分数： {}分後".format(num))
```

## 汎用的に利用する関数

引数を(倍々に増えていく物体、覆われるものの体積、倍になる分数)の3つを設定し、  
bainbain関数を作成

```python
def bainbain(volume_bf=100,volume_af=6.600827495946115e+61,bai=5):
    num = 0
    while volume_bf <= volume_af:
        #print("経過： {}分　体積： {}".format(num,volume_af))
        volume_bf *= 2
        num += bai
    else:
        print("いっぱいになる分数： {}分後".format(num))
```

```python
bainbain() #関数を利用し栗饅頭が太陽系を覆う分数を計算
```

## 汎用的に利用する関数からリストを作成

上記で作成したbainbain関数を活用し、その中で増えていく体積のリストと  
経過する分数のリストの2つを作成するbainbain2関数を作成

```python
def bainbain2(volume_bf=100,volume_af=6.600827495946115e+61,bai=5):
    num = 0
    volume_list =[] #増える体積のリスト
    num_list = [] #経過分数のリスト
    
    while volume_bf <= volume_af:
        #print("経過： {}分　体積： {}".format(num,volume_af))        
        volume_bf *= 2
        num += bai
        volume_list.append(volume_bf)
        num_list.append(num)
    else:
        print("いっぱいになる分数： {}分後".format(num))
        #print(volume_list)
        #print(num_list)
    
    return volume_list,num_list
volume_list,num_list = bainbain2(volume_bf=100,volume_af=6.600827495946115e+61,bai=5)
```

```python
bainbain2() #栗まんじゅうでリスト作成
```

## bainbain2関数で作成した2つのリストを利用し、経過分数ごとの増加体積をグラフ化

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("bainbain")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(num_list,volume_list) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```
