## 【問題1】乱数の作成

```python
import numpy as np
from numpy.random import *
from matplotlib import pyplot as plt
import seaborn as sns

mu = [-3, 0]
sigma = [[1.0, 0.8], [0.8, 1.0]]
 
np.random.seed(0)
    
# 2次元正規乱数を500個生成
values = np.random.multivariate_normal(mu, sigma, 500)
print (values)

```

```python
sns.jointplot(values[:,0], values[:,1])
plt.show()
```

## 【問題2】散布図による可視化

```python
import matplotlib.pyplot as plt
plt.title("scatter")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(values[:,0], values[:,1],c ="r",alpha=0.3)
plt.show()
```

## 【問題4】データの追加

```python
mu_2 = [0, -3]
sigma_2 = [[1.0, 0.8], [0.8, 1.0]]

values_2 = np.random.multivariate_normal(mu_2, sigma_2, 500)
#print (values_2)

np.random.seed(0)
```

## 【問題5】データの結合
