### 初期の重み設定

#### 全結合

```python
class SimpleInitializer:
    """
    ガウス分布によるシンプルな初期化
    Parameters
    ----------
    sigma : float
      ガウス分布の標準偏差
    """
    def __init__(self, sigma):
        self.sigma = sigma
        
    def W(self, n_nodes1, n_nodes2):
        """
        重みの初期化
        Parameters
        ----------
        n_nodes1 : int
          前の層のノード数
        n_nodes2 : int
          後の層のノード数

        Returns
        ----------
        W :
        """
        W= self.sigma * np.random.randn(n_nodes1, n_nodes2)
        return W
    
    def B(self, n_nodes2):
        """
        バイアスの初期化
        Parameters
        ----------
        n_nodes2 : int
          後の層のノード数

        Returns
        ----------
        B :
        """
        B =  np.random.randn(n_nodes2,)
        return B
```

```python
class CONV:
    """
    ノード数n_nodes1からn_nodes2への全結合層
    Parameters
    ----------
    n_nodes1 : int
      前の層のノード数
    n_nodes2 : int
      後の層のノード数
    initializer : 初期化方法のインスタンス
    optimizer : 最適化手法のインスタンス
    """
    def __init__(self,N_in,Pading,fil_num,Slide,fil_size,initializer, optimizer):
        #self.optimizer = optimizer
        # 指定の重みとバイアスを代入
        # initializerのメソッドを使い、self.Wとself.Bを初期化する
        self.W=initializer.W(fil_num, fil_size)
        self.B =initializer.B(fil_num,)
        self.N_in = N_in
        self.Pading = Pading
        self.Fil = fil_size
        self.Slide = Slide
    
    def output_size(self,):
        N_ount = (self.N_in+2*self.Pading-self.Fil/self.Slide)+1
        return N_ount


    def forward(self, X):
        """
        フォワード
        Parameters
        ----------
        X : 次の形のndarray, shape (batch_size, n_nodes1)
            入力
        Returns
        ----------
        A : 次の形のndarray, shape (batch_size, n_nodes2)
            出力
        """        
        self.X = X
        print('ConvX',self.X.shape)
        
        #アウトプットサイズ
        self.N_h_out = self.output_size()
        self.N_h_out_int = np.array(self.N_h_out).astype(np.int)

        self.N_w_out = self.output_size()
        self.N_w_out_int = np.array(self.N_w_out).astype(np.int)

        
        #a_1d=np.arange(N_h_out_int,N_h_out_int)
        loss_out_arr_cul= []
        #最終格納するリスト 追加
        a_2d_ijm = np.empty((0,self.N_h_out_int), int)

        #フィルタ分繰り返すを追加
        for m in range (w_2d_re.shape[2]):
            loss_out_arr_cul= []
            for i in range(self.N_w_out_int ):
                for  j in range(self.N_h_out_int):
                    a = 0
                    ab = 0
                    for s in range(w_2d_re.shape[1]):
                        for t in range(w_2d_re.shape[0]):
                            a += self.X[0,i+s, j+t] * w_2d_re[m,s,t] 
                    ab = a+self.B[m]
                    loss_out_arr_cul.append(ab)
            a_2d_re = np.reshape(loss_out_arr_cul,[self.N_w_out_int ,self.N_h_out_int]) #26,26
            a_2d_ijm = np.append(a_2d_ijm,np.array(a_2d_re), axis=0)
            #print('loss_out_arr_cul',len(loss_out_arr_cul))
        a_2d_ijm = np.reshape(a_2d_ijm,[w_2d_re.shape[2],self.N_w_out_int ,self.N_h_out_int])
        #print('a_2d_ijm',a_2d_ijm.shape)

        return a_2d_ijm
      
    def backward(self, delta_a):
        """
        バックワード
        Parameters
        ----------
        dA : 次の形のndarray, shape (batch_size, n_nodes2)
            後ろから流れてきた勾配
        Returns
        ----------
        dZ : 次の形のndarray, shape (batch_size, n_nodes1)
            前に流す勾配
        """
 
        #バイアスをa_2d_ijmから算出
        self.B_new = np.sum(a_2d_ijm,axis = 0)
        
        #重み（フィルタ）の結果格納のリスト
        w_new_2d =[]
        w_new_pre = 0

        #最終格納するリスト 追加
        w_2d_ijm = np.empty((0,w_2d_re.shape[0]), int)


        for m in range(a_2d_ijm.shape[2]):
            w_new_2d =[]
            for t in range(w_2d_re.shape[0]):
                for s in range(w_2d_re.shape[1]):
                    w_new_pre = 0
                    for j in range(N_h_out_int):
                        for i in range(N_h_out_int):
                            w_new_pre += a_2d_ijm[i,j,m] * X_2d[0,0,i+s, j+t]
                    w_new_2d.append(w_new_pre)
            w_2d_new = np.reshape(w_new_2d,[w_2d_re.shape[0],w_2d_re.shape[1]])
            w_2d_ijm = np.append(w_2d_ijm,np.array(w_2d_new), axis=0)
            #print('w_new_2d',len(w_new_2d))
        self.W_new = np.reshape(w_2d_ijm,[w_2d_re.shape[0],w_2d_re.shape[1],w_2d_re.shape[2]])
        #print('w_2d_ijm',w_2d_ijm.shape)


            
        #Xの更新
        #Xのアウトプットのリスト
        x_new_2d =[]

        #最終格納するリスト 追加
        x_2d_ijm = np.empty((0,w_2d_re.shape[0]), int)


        for m in range(a_2d_ijm.shape[2]):
            x_new_2d =[]
            for i in range(X_2d.shape[2]):
                delta_x = 0
                for j in range(X_2d.shape[3]):
                    for s in range(w_2d.shape[0]):
                        for t in range(w_2d.shape[1]):
                            i_s=i-s
                            j_t=j-t
                        if i_s < 0:
                            delta_x += 0*w_2d_re[s,t,m]
                        else:
                            if  i_s > N_h_out_int-1:
                                delta_x += 0*w_2d_re[s,t,m]
                            else:
                                if j_t < 0:
                                    delta_x += 0*w_2d_re[s,t,m]
                                else:
                                    if  j_t > N_w_out_int-1: 
                                        delta_x += 0*w_2d_re[s,t,m]
                                    else:
                                        delta_x += a_2d_ijm[i_s,j_t,m]*w_2d_re[s,t,m]
                    x_new_2d.append(delta_x)
                    #print('x_new_2d',len(x_new_2d))
            x_2d_new += np.reshape(x_new_2d,[X_2d.shape[2],X_2d.shape[3]])
            #print('x_2d_new',x_2d_new.shape)
    
    
        # 更新
        self = self.optimizer.update(self)
        
        return x_2d_new
```

### 更新式

```python
class SGD:
    """
    確率的勾配降下法
    Parameters
    ----------
    lr : 学習率
    """
    def __init__(self, lr):
        self.lr = lr
    def update(self, layer):
        """
        ある層の重みやバイアスの更新
        Parameters
        ----------
        layer : 更新前の層のインスタンス
        """
        layer.W -= self.lr*layer.W_new
        layer.B -=  self.lr*layer.B_new
```

### 全結合

```python
class FC:
    """
    ノード数n_nodes1からn_nodes2への全結合層
    Parameters
    ----------
    n_nodes1 : int
      前の層のノード数
    n_nodes2 : int
      後の層のノード数
    initializer : 初期化方法のインスタンス
    optimizer : 最適化手法のインスタンス
    """
    def __init__(self, n_nodes1, n_nodes2,initializer, optimizer):
        self.optimizer = optimizer
        # 初期化
        # initializerのメソッドを使い、self.Wとself.Bを初期化する
        self.W=initializer.W(n_nodes1, n_nodes2)
        self.B =initializer.B(n_nodes2)
        
    def forward(self, X):
        """
        フォワード
        Parameters
        ----------
        X : 次の形のndarray, shape (batch_size, n_nodes1)
            入力
        Returns
        ----------
        A : 次の形のndarray, shape (batch_size, n_nodes2)
            出力
        """        
        self.X = X
        A = np.dot(X,self.W)+self.B
        
        return A
    
    def backward(self, dA):
        """
        バックワード
        Parameters
        ----------
        dA : 次の形のndarray, shape (batch_size, n_nodes2)
            後ろから流れてきた勾配
        Returns
        ----------
        dZ : 次の形のndarray, shape (batch_size, n_nodes1)
            前に流す勾配
        """
        self.B_new = np.sum(dA,axis = 0)
        self.W_new = np.dot(self.X.T,dA)
        dZ = np.dot(dA,self.W.T)

        # 更新
        self = self.optimizer.update(self)
        
        return dZ
```

### 活性化関数

```python
class Tanh():
    
    def __init__(self):
        #self.optimizer = optimizer
        # 初期化
        # initializerのメソッドを使い、self.Wとself.Bを初期化する
        pass
    
    def forward(self, X):
        """
        フォワード
        Parameters
        ----------
        X : 次の形のndarray, shape (batch_size, n_nodes1)
            入力
        Returns
        ----------
        A : 次の形のndarray, shape (batch_size, n_nodes2)
            出力
        """        
        self.X = X
        self.A =np.tanh(X)
        
        return self.A
    
    def backward(self, dZ):
        """
        バックワード
        Parameters
        ----------
        dA : 次の形のndarray, shape (batch_size, n_nodes2)
            後ろから流れてきた勾配
        Returns
        ----------
        dZ : 次の形のndarray, shape (batch_size, n_nodes1)
            前に流す勾配
        """
        dA = dZ*(1-np.tanh(self.A)**2)

        return dA
```

### ソフトマックス

```python
class Softmax():
    
    def __init__(self):
        #self.optimizer = optimizer
        # 初期化
        # initializerのメソッドを使い、self.Wとself.Bを初期化する
        self.loss_list = []
        pass
    
    def forward(self, X):
        """
        フォワード
        Parameters
        ----------
        X : 次の形のndarray, shape (batch_size, n_nodes1)
            入力
        Returns
        ----------
        Z : 次の形のndarray, shape (batch_size, n_nodes2)
            出力
        """        
        #20サンプルのソフトマックス分子作成
        bunbo = np.sum(np.exp(X),axis=1)
        #20サンプル、10クラスと割り算できるようにリシェーぷ
        bunboT = np.reshape(bunbo,[X.shape[0],1])
        #ソフトマックス計算でZ3算出
        Z = np.exp(X)/bunboT
  
        return Z
    
    def backward(self, dZ,Y):
        #バックワード
        dA = dZ-Y
        #交差エントロピー
        loss = -np.sum(Y*np.log(dZ))/Y.shape[0]
        self.loss_list.append(loss)
        
        return dA,self.loss_list
```

### ミニバッチ

```python
class GetMiniBatch:
    """
    ミニバッチを取得するイテレータ

    Parameters
    ----------
    X : 次の形のndarray, shape (n_samples, n_features)
      訓練用データ
    y : 次の形のndarray, shape (n_samples, 1)
      正解値
    batch_size : int
      バッチサイズ
    seed : int
      NumPyの乱数のシード
    """
    def __init__(self, X, y, batch_size = 20, seed=0):
        self.batch_size = batch_size
        np.random.seed(seed)
        shuffle_index = np.random.permutation(np.arange(X.shape[0]))
        self._X = X[shuffle_index]
        self._y = y[shuffle_index]
        self._stop = np.ceil(X.shape[0]/self.batch_size).astype(np.int)

    def __len__(self):
        return self._stop

    def __getitem__(self,item):
        p0 = item*self.batch_size
        p1 = item*self.batch_size + self.batch_size
        return self._X[p0:p1], self._y[p0:p1]        

    def __iter__(self):
        self._counter = 0
        return self

    def __next__(self):
        if self._counter >= self._stop:
            raise StopIteration()
        p0 = self._counter*self.batch_size
        p1 = self._counter*self.batch_size + self.batch_size
        self._counter += 1
        return self._X[p0:p1], self._y[p0:p1]
```

```python
class Scratch2dCNNClassifier():
    
    def __init__(self, iter_num = 100,verbose = True,α=0.00001,n_features = 28,n_nodes1 = 128
                 ,sigma = 0.01,n_nodes2 = 64,n_output = 10,batchi_sumple=20,fil_num_1 = 32,fil_num_2=64,
                 fil_num_3=128,fil_size=3,fil_slide=1,max_fil_size=2,max_fil_slide= 2,Pading = 0,max_out_1=13,max_out_2=5,max_out_3=1):
        self.verbose = verbose
        self.lr = α
        self.n_features =n_features
        self.n_nodes1 = n_nodes1
        self.sigma = sigma
        self.n_nodes2=n_nodes2
        self.n_output=n_output
        self.batchi_sumple =batchi_sumple
        self.iter_num = iter_num
        self.fil_num_1 =fil_num_1
        self.fil_num_2 =fil_num_2
        self.fil_num_3 =fil_num_3
        self.fil_size=fil_size
        self.fil_slide=fil_slide
        self.max_fil_size=max_fil_size
        self.max_fil_slide=max_fil_slide
        self.Pading = Pading
        self.max_out_1=max_out_1
        self.max_out_2=max_out_2
        self.max_out_3=max_out_3
        
        pass
    
    
    def fit(self, X, y, X_val=None, y_val=None):
        #インスタンス化
        optimizer = SGD(self.lr)
              
        #CNNコンポ1
        self.CONV1 = CONV(self.n_features, self.Pading,self.fil_num_1, self.fil_slide,self.fil_size,SimpleInitializer_con(self.sigma), optimizer)
        self.MAX1 = MAX_pool(self.max_fil_size,self.max_fil_size,self.max_out_1)
       #CNNコンポ2
        self.CONV2 = CONV(self.n_features, self.Pading,self.fil_num_2, self.fil_slide,self.fil_size,SimpleInitializer_con(self.sigma), optimizer)
        self.MAX2 = MAX_pool(self.max_fil_size,self.max_fil_size,self.max_out_2)
       #CNNコンポ3
        self.CONV3 = CONV(self.n_features, self.Pading,self.fil_num_3, self.fil_slide,self.fil_size,SimpleInitializer_con(self.sigma), optimizer)
        self.MAX3 = MAX_pool(self.max_fil_size,self.max_fil_size,self.max_out_3)
        
        #フラット
        self.FLAT=FLAT()
        
        #全結合
        #self.FC1 = FC(self.n_features, self.n_nodes1, SimpleInitializer(self.n_nodes1), optimizer)
        #self.activation1 = Tanh()
        self.FC2 = FC(self.n_nodes1, self.n_nodes2, SimpleInitializer(self.sigma), optimizer)
        self.activation2 = Tanh()
        self.FC3 = FC(self.n_nodes2, self.n_output, SimpleInitializer(self.sigma), optimizer)
        self.activation3 = Softmax()
        
        #データを取得
        get_mini_batch = GetMiniBatch(X, y, batch_size=self.batchi_sumple)
        #ロスのリスト作成
        loss_list = []
        
        for i in range(self.iter_num):
            for mini_X_train, mini_y_train in get_mini_batch:
                #フォワード
                #A1 = self.FC1.forward(mini_X_train)
                #Z1 = self.activation1.forward(A1)
                
                C1 = self.CONV1.forward(mini_X_train)
                M1 = self.MAX1.forward(C1)
                C2 = self.CONV2.forward(M1)
                M2 = self.MAX2.forward(C2)
                C3 = self.CONV3.forward(M2)
                M3 = self.MAX3.forward(C3)
                
                #フラット
                F1 = self.FLAT.forward(M3)
                
                A2 = self.FC2.forward(F1)
                Z2 = self.activation2.forward(A2)
                A3 = self.FC3.forward(Z2)
                Z3 = self.activation3.forward(A3)
                
                #バックワード
                dA3,loss_list = self.activation3.backward(Z3, mini_y_train) # 交差エントロピー誤差とソフトマックスを合わせている
                dZ2 = self.FC3.backward(dA3)
                dA2 = self.activation2.backward(dZ2)
                dZ1 = self.FC2.backward(dA2)
                #dA1 = self.activation1.backward(dZ1)
                #dZ0 = self.FC1.backward(dA1) 
                
                dF1 = self.FLAT.backward(dZ1)
                
                dM3 = self.CONV3.backward(dF1)
                dC3 = self.CONV3.forward(dM3)
                dM2 = self.CONV2.forward(dC3)
                dC2 = self.CONV2.forward(dM2)
                dM1 = self.CONV1.forward(dC2)
                dC1 = self.CONV1.forward(dM1)                
                             
        return loss_list
    
    def predict(self, X):
         #フォワード
        A1 = self.FC1.forward(X)
        Z1 = self.activation1.forward(A1)
        A2 = self.FC2.forward(Z1)
        Z2 = self.activation2.forward(A2)
        A3 = self.FC3.forward(Z2)
        Z3 = self.activation3.forward(A3)

        pred = np.argmax(Z3,axis = 1)

        return pred
```
